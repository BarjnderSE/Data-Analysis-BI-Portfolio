# Tips Sales data visualization with Mataplotlib library

```python
# 1. Import necessary libraries
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

import warnings

import warnings

# Ignore all future warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
```

```python
# load the dataset 
tips = sns.load_dataset('tips')
print(tips.head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Load the Tips dataset
tips = sns.load_dataset('tips')

# Calculate average sales by day
avg_sales = tips.groupby('day')['total_bill'].mean()

# Create a bar plot
plt.figure(figsize=(8, 6))
bars = avg_sales.plot(kind='bar', color='skyblue')

# Add labels on top of bars with adjusted position
for i in range(len(avg_sales)):
    plt.text(i, avg_sales[i] + 0.5, f'{avg_sales[i]:.2f}', ha='center', va='bottom', color='black', fontsize=12)
    
plt.title('Average Sales by Day')
plt.xlabel('Day')
plt.ylabel('Average Total Bill (Sales)')
plt.xticks(rotation=45)
plt.show()
```

# Explanation:

    We grouped the data by day and calculated the average sales (total_bill) for each day.
    The bar plot helps to compare how the sales vary across different days of the week.

# 2. Line plot for sales over time/ How the total bill changes over the course of the week (i.e., by day) to identify any trends or patterns?

```python


# Load the Tips dataset
tips = sns.load_dataset('tips')

# Calculate average total bill by day
avg_sales = tips.groupby('day')['total_bill'].mean()

# Create a line plot
plt.figure(figsize=(10,6))

# Plot the line
plt.plot(avg_sales.index, avg_sales.values, marker='o', color='skyblue')

# Add labels on top of the line
for i in range(len(avg_sales)):
    plt.text(i, avg_sales[i] + 0.5, f'{avg_sales[i]:.2f}', ha='center', va='bottom', color='black', fontsize=12)

# Add title and labels
plt.title('Average Sales by Day')
plt.xlabel('Day')
plt.ylabel('Average Sales')

# Adjust layout to prevent clipping of labels
plt.tight_layout()

# Display the plot
plt.show()
```

# observation from above line chart

    The line chart shows the average sales by day, which provides a clear picture of how the total bill changes over the course of the week.
    The line plot also allows us to identify any trends or patterns in the data, such as a spike in sales on a particular day or a consistent increase in sales over the week.

# 8 Create Product Analysis for Tips Dataset using Correlation Matrix and Heatmap

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Tips dataset
tips = sns.load_dataset('tips')

# Select only the numeric columns
numeric_tips = tips.select_dtypes(include=['float64', 'int64'])

# Calculate the correlation matrix for numeric columns
corr_matrix = numeric_tips.corr()

# Create a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix of Numeric Columns')
plt.show()

```

Key Observations from the Correlation Matrix and Heatmap:

    Total Bill and Tip: A moderate positive correlation of 0.68 indicates that as the total bill increases, the tip also increases, as expected with percentage-based tipping.

    Total Bill and Size: A weak positive correlation suggests larger groups generally have higher total bills, but the relationship isn't as strong as with tips.

    Size and Tip: Weak or negligible correlation, indicating that tip amounts are more closely related to the total bill than to the number of people in the group.

    No Strong Negative Correlations: There are no significant negative correlations among numeric variables, showing no inverse relationships.

    Categorical Variables: Non-numeric columns like 'sex' or 'smoker' were excluded from the correlation, but could be included after encoding.

    Heatmap Insights: The heatmap visually highlights strong correlations, with darker colors indicating stronger relationships, making it easy to identify patterns for further analysis.

# 9 Box plot to analyze the distribution of total bills by day of the week and time of day


```python
plt.figure(figsize=(10,6))

sns.boxplot(x='day', y='total_bill', hue='time', data=tips, palette='coolwarm')

plt.title('Total Bill Distribution by Day and Time')

plt.xlabel('Day')

plt.ylabel('Total Bill')

plt.show()
```

# observation from above box plot

Key Observations from the Box Plot:

    Higher Total Bills for Dinner: The box plot clearly shows that the median total bill is higher for Dinner (in orange) compared to Lunch (in blue) on all days of the week, indicating that people tend to spend more during dinner.

    Variation in Total Bills: On Thursday and Friday, the range of total bills for both lunch and dinner is wider, suggesting more variation in spending. Particularly on Thursday, the bills during dinner show significant outliers.

    Weekends (Saturday and Sunday): For both Lunch and Dinner, the total bills are more consistent on Saturday and Sunday, with fewer outliers compared to Thursday and Friday.

    Outliers: Several extreme values are observed across different days, particularly for Dinner on Thursday and Friday, indicating that some bills are significantly higher than the rest.

    Day-by-Day Comparison: Overall, Dinner tends to have higher total bills, but the variation across days shows that the spending habits differ depending on both the day of the week and the meal time.

# 10 Area Chart 
# How does the total bill vary across days of the week for smokers and non-smokers?"



```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create an area chart of total bill by day and smoker status
plt.figure(figsize=(10,6))

sns.lineplot(x='day', y='total_bill', hue='smoker', data=tips, palette='coolwarm', markers='o', ci=None)

# Adding shaded areas to the plot for an area-like effect
plt.fill_between(tips[tips['smoker'] == 'Yes'].groupby('day')['total_bill'].mean().index,
                 tips[tips['smoker'] == 'Yes'].groupby('day')['total_bill'].mean(), alpha=0.3)
plt.fill_between(tips[tips['smoker'] == 'No'].groupby('day')['total_bill'].mean().index,
                 tips[tips['smoker'] == 'No'].groupby('day')['total_bill'].mean(), alpha=0.3)

plt.title('Total Bill Distribution by Day and Smoker Status')
plt.xlabel('Day')
plt.ylabel('Total Bill')
plt.show()

```

The area chart shows that smokers tend to have higher total bills than non-smokers, especially on Friday and Saturday,Sunday while the difference is smaller on Thursday This suggests smokers may spend more on dining during weekends. To further analyze, we could explore factors such as group size, time of day, or specific menu items to understand the drivers behind these spending patterns. Additionally, marketing efforts or promotions targeting on weekdays could be considered to balance the spending across the week.

# 11. Subplots to visualize the total_bill,average_sales,tip, and size distribution.

```python


# Calculate average sales by day
avg_sales = tips.groupby('day')['total_bill'].mean()

# Calculate sales by time of day (Lunch/Dinner)
sales_by_time = tips.groupby('time')['total_bill'].sum()

# Create subplots (2 rows, 2 columns)
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# Bar plot for average sales by day
axes[0, 0].bar(avg_sales.index, avg_sales.values, color='skyblue')
axes[0, 0].set_title('Average Sales by Day')
axes[0, 0].set_xlabel('Day')
axes[0, 0].set_ylabel('Sales')

# Histogram for total bill
axes[0, 1].hist(tips['total_bill'], bins=10, color='purple', edgecolor='black')
axes[0, 1].set_title('Distribution of Sales')
axes[0, 1].set_xlabel('Total Bill (Sales)')
axes[0, 1].set_ylabel('Frequency')

# Scatter plot for sales vs tip
axes[1, 0].scatter(tips['total_bill'], tips['tip'], color='blue')
axes[1, 0].set_title('Sales vs Tip')
axes[1, 0].set_xlabel('Total Bill (Sales)')
axes[1, 0].set_ylabel('Tip (Profit)')

# Pie chart for sales by time of day
axes[1, 1].pie(sales_by_time, labels=sales_by_time.index, autopct='%1.1f%%', colors=['lightblue', 'orange'])
axes[1, 1].set_title('Sales Distribution by Time of Day')

# Adjust layout to avoid overlap
plt.tight_layout()

# Show the plot
plt.show()
```
