```python
import pandas as pd
```

```python



# Create the dataset 
data = {
    'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'],
    'Smartphone Sales': [150, 120, 170, 180, 160, 190, 210, 220, 230, 240, 250, 300],
    'Laptop Sales': [200, 210, 230, 220, 240, 250, 260, 270, 280, 290, 300, 350],
    'Tablet Sales': [100, 130, 140, 160, 170, 180, 200, 210, 230, 250, 260, 300]
}

# Create a DataFrame
df = pd.DataFrame(data)

# Display the DataFrame
print(df)

```

# 1.Line Plot: Create a line plot to visualize the sales of Smartphones, Laptops, and Tablets over the months. What trends do you notice for each product?

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(12,4))
plt.plot(df['Month'],df['Smartphone Sales'],marker='o',label='Smartphone')
plt.plot(df['Month'],df['Laptop Sales'],marker='o',label='Laptops')
plt.plot(df['Month'],df['Tablet Sales'],marker='o',label='Tablets')


plt.title('Monthly sales of products')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.legend()
plt.grid()
plt.show()
```

# 2. Bar Chart Question: Generate a bar chart showing the total sales of each product.

```python
total_sales=df[['Smartphone Sales','Laptop Sales', 'Tablet Sales']].sum()

total_sales.plot(kind='bar',figsize=(8,5))
plt.title('Total sales of each Product')
plt.ylabel('Total Sales')
plt.xlabel('Product')
plt.xticks(rotation=45)
plt.show()
```

# 3. Pie Chart Question: Create a pie chart representing the total sales distribution of each product.

```python
total_sales = df[['Smartphone Sales', 'Laptop Sales', 'Tablet Sales']].sum()

# Explode the Laptop slice
explode = (0, 0.1, 0)  # Only "explode" the second slice (Laptop Sales)

# Plotting
plt.figure(figsize=(8, 8))
plt.pie(total_sales, labels=total_sales.index, autopct='%1.1f%%', startangle=140, explode=explode)
plt.title('Sales Distribution by Product')
plt.show()
```

# 4. Scatter Plot Question: Create a scatter plot comparing the sales of Smartphones and Laptops.

```python
plt.figure(figsize=(8, 5))
plt.scatter(df['Smartphone Sales'], df['Laptop Sales'])
plt.title('Smartphone Sales vs Laptop Sales')
plt.xlabel('Smartphone Sales')
plt.ylabel('Laptop Sales')
plt.grid()
plt.show()
```

# 5. Box Plot Question: Generate a box plot for the sales of each product.

```python
plt.figure(figsize=(10, 6))
df[['Smartphone Sales', 'Laptop Sales', 'Tablet Sales']].boxplot()
plt.title('Sales Distribution by Product')
plt.ylabel('Sales')
plt.show()
```

# 6. Histogram Question: Create a histogram showing the frequency of sales for Tablets.

```python
plt.figure(figsize=(8, 5))
plt.hist(df['Tablet Sales'], bins=5, edgecolor='black')
plt.title('Histogram of Tablet Sales')
plt.xlabel('Sales')
plt.ylabel('Frequency')
plt.show()
```

# 7. Monthly Growth Question: Calculate the month-over-month growth rate for each product and plot it.

```python
growth = df[['Smartphone Sales', 'Laptop Sales', 'Tablet Sales']].pct_change() * 100

# Plotting
growth.plot(kind='bar', figsize=(10, 6))
plt.title('Monthly Growth Rate (%)')
plt.ylabel('Growth Rate (%)')
plt.xlabel('Month')
plt.xticks(rotation=45)
plt.show()
```

# 8. Comparison of Best Months Question: Identify which month had the highest sales for each product.

```python
best_months = df.loc[df[['Smartphone Sales', 'Laptop Sales', 'Tablet Sales']].idxmax()]
print(best_months[['Month', 'Smartphone Sales', 'Laptop Sales', 'Tablet Sales']])
```
